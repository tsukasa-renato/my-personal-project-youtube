```python
import googleapiclient.discovery
from utils.util import str_clear

my_api_key = 'Your Key'
api_service_name = 'youtube'
api_version = 'v3'

youtube = googleapiclient.discovery.build(
    api_service_name, api_version, developerKey=my_api_key)
```

```python
def collect_comments(video_id, page_token):
    """
    This function list comments of a youtube video

    :param video_id: id of the youtube video, after ?v= -> youtube.com/watch?v=video_id
    :param page_token:
    :return: return 100 comments
    """

    results = youtube.commentThreads().list(
        part="snippet,replies",
        maxResults=100,
        videoId=video_id,
        textFormat="plainText",
        pageToken=page_token
    ).execute()

    print('100 comments returned')

    return results


def write_json_file(comments, name):
    """
    This function saves comments in a txt file

    :param comments: collected comments
    :param name: name for the txt file
    :return: a txt file
    """
    
    text = ''
    
    for item in comments:

        comment = item["snippet"]["topLevelComment"]["snippet"]
        
        text += '{'
        text += '"' + 'is_replies":' + '"' + 'False",'
        text += '"' + 'author":' + '"' + str_clear(comment['authorDisplayName']) + '",'
        text += '"' + 'comment":' + '"' + str_clear(comment["textDisplay"]) + '",'
        text += '"' + 'date":' + '"' + str_clear(comment["updatedAt"]) + '",'
        text += '"' + 'like":' + '"' + str(comment["likeCount"]) + '",'
        text += '},'
        
        if 'replies' in item.keys():
            
            replies = item["replies"]["comments"]
            
            for reply in replies:
                text += '{'
                text += '"' + 'is_replies":' + '"' + 'True",'
                text += '"' + 'author":' + '"' + str_clear(reply["snippet"]['authorDisplayName']) + '",'
                text += '"' + 'comment":' + '"' + str_clear(reply["snippet"]["textDisplay"]) + '",'
                text += '"' + 'date":' + '"' + str_clear(reply["snippet"]["updatedAt"]) + '",'
                text += '"' + 'like":' + '"' + str(reply["snippet"]["likeCount"]) + '",'
                text += '},'

    with open('database/comments/' + name + '.json', '+a', encoding="utf8") as file:
        file.write(text)

    print('Information was stored in txt file')
```

```python
import pandas as pd

name_file = 'ciro gomes'

videos = pd.read_json(f'database/videos/{name_file}.json')
videos = videos.ID.tolist()
```

```python
for video in videos:
    
    try:
        
        comment_data = collect_comments(video, '')
        
        with open('database/comments/' + video + '.json', '+a', encoding="utf8") as file:
            file.write('[')
        
        while True:
            
            write_json_file(comment_data['items'], video)
            comment_data = collect_comments(video, comment_data['nextPageToken'])
            
    except KeyError:
        
        with open('database/comments/' + video + '.json', '+a', encoding="utf8") as file:
            file.write(']')
        print('Comments successfully collected')
```
